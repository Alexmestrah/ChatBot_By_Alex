```python
from tkinter import *
from tkinter import ttk
from tkinter import PhotoImage
from PIL import Image, ImageTk
import json
import testing_model as test_py  
import training_model as train_py  

# Define color constants
BG_GRAY = "#ABB2B9"
BG_COLOR = "#34495E"  # Updated background color for a modern look
TEXT_COLOR = "#FFF"
FONT = "Helvetica 12"  # Reduced font size for better readability
FONT_BOLD = "Helvetica 12 bold"  # Bold font for headings

class ChatBot:
    def __init__(self):
        self.window = Tk()
        self.main_window()
        self.window.iconbitmap('icon_alex_logo.ico') 
        self.test = test_py.Testing()

    def run(self):
        self.window.mainloop()

    def main_window(self):
        self.window.title("ChatBot")
        self.window.geometry("820x720")  # Set window size
        self.window.configure(bg=BG_COLOR)
        
        # Add tabs for Chatbot and Train Bot in Notebook frame with updated styling
        self.tab = ttk.Notebook(self.window)
        self.tab.pack(expand=1, fill='both')
        self.bot_frame = ttk.Frame(self.tab)
        self.train_frame = ttk.Frame(self.tab)
        self.tab.add(self.bot_frame, text='Chatbot')
        self.tab.add(self.train_frame, text='Entraîner le Bot')

        # Add elements to the Chatbot tab
        self.add_bot()
        self.add_train()

    def add_bot(self):
        # Add heading to the Chatbot window
        head_label = Label(self.bot_frame, bg=BG_COLOR, fg=TEXT_COLOR,
                           text="Bienvenue sur le CV d'Alex", font=FONT_BOLD, pady=10)
        head_label.pack(fill='x')

        # Create text widget where conversation will be displayed with updated styling
        self.text_widget = Text(self.bot_frame, width=50, height=20, bg="#fff", fg="#000", font=FONT, padx=5, pady=5)
        self.text_widget.pack(pady=10, padx=10, fill='both', expand=True)
        self.text_widget.configure(state=DISABLED)

        # Create scrollbar
        scrollbar = Scrollbar(self.text_widget)
        scrollbar.pack(side=RIGHT, fill=Y)
        scrollbar.config(command=self.text_widget.yview)
        self.text_widget.config(yscrollcommand=scrollbar.set)

        # Create bottom label where message widget will be placed
        bottom_label = Frame(self.bot_frame, bg=BG_COLOR)
        bottom_label.pack(side=BOTTOM, fill='x')

        # Entry widget for user input with updated styling
        self.msg_entry = Entry(bottom_label, bg="#fff", fg="#000", font=FONT)
        self.msg_entry.pack(side=LEFT, fill='x', expand=True, padx=10, pady=10)
        self.msg_entry.focus()
        self.msg_entry.bind("<Return>", self.on_enter)

        # Send button to send user message with updated styling
        send_button = Button(bottom_label, text="Envoyer", font=FONT_BOLD, bg="#2ECC71", fg="#FFF",
                             command=lambda: self.on_enter(None))
        send_button.pack(side=RIGHT, padx=10, pady=10)

    def add_train(self):
        # Add heading to the Train Bot window
        head_label = Label(self.train_frame, bg=BG_COLOR, fg=TEXT_COLOR, text="Entraîner le Bot", font=FONT_BOLD, pady=10)
        head_label.pack(fill='x')

        # Tag Label and Entry for data tag
        tag_label = Label(self.train_frame, fg="#000", text="Tag", font=FONT)
        tag_label.pack(anchor='w', padx=10)
        self.tag = Entry(self.train_frame, bg="#fff", fg="#000", font=FONT)
        self.tag.pack(fill='x', padx=10, pady=5)

        # Pattern Label and Entry for pattern to our tag with updated styling
        self.pattern = []
        for i in range(4):
            pattern_label = Label(self.train_frame, fg="#000", text="Pattern %d" % (i + 1), font=FONT)
            pattern_label.pack(anchor='w', padx=10)
            pattern_entry = Entry(self.train_frame, bg="#fff", fg="#000", font=FONT)
            pattern_entry.pack(fill='x', padx=10, pady=5)
            self.pattern.append(pattern_entry)

        # Response Label and Entry for response to our pattern with updated styling
        self.response = []
        for i in range(2):
            response_label = Label(self.train_frame, fg="#000", text="Réponse %d" % (i + 1), font=FONT)
            response_label.pack(anchor='w', padx=10)
            response_entry = Entry(self.train_frame, bg="#fff", fg="#000", font=FONT)
            response_entry.pack(fill='x', padx=10, pady=5)
            self.response.append(response_entry)

        # Train button to train the bot
        train_button = Button(self.train_frame, text="Entraîner le Bot", font=FONT_BOLD, bg="#3498DB", fg="#FFF",
                              command=lambda: self.on_train(None))
        train_button.pack(pady=20)

    def on_train(self, event):
        # Read data file and append created tag, pattern, and responses from add_train function
        with open('data.json', 'r+') as json_file:
            file_data = json.load(json_file)
            file_data['intents'].append({
                "tag": self.tag.get(),
                "patterns": [i.get() for i in self.pattern],
                "réponses": [i.get() for i in self.response],
                "context": ""
            })
            json_file.seek(0)
            json.dump(file_data, json_file, indent=1)
        # Run and compile model from our training_model.py file.
        train = train_py.Training()
        train.build()
        print("Trained Successfully")
        self.test = test_py.Testing()

    def on_enter(self, event):
        # Get user query and bot response
        msg = self.msg_entry.get()
        self.my_msg(msg, "You")
        self.bot_response(msg, "Bot")

    def bot_response(self, msg, sender):
        self.text_widget.configure(state=NORMAL)
        # Get the response for the user's query from testing_model.py file
        self.text_widget.insert(END, str(sender) + " : " + self.test.response(msg) + "\n\n")
        self.text_widget.configure(state=DISABLED)
        self.text_widget.see(END)

    def my_msg(self, msg, sender):
        # It will display user query and bot response in text_widget
        if not msg:
            return
        self.msg_entry.delete(0, END)
        self.text_widget.configure(state=NORMAL)
        self.text_widget.insert(END, str(sender) + " : " + str(msg) + "\n")
        self.text_widget.configure(state=DISABLED)


# Run the file
if __name__ == "__main__":
    bot = ChatBot()
    bot.run()
```
